```python
import numpy as np
import thinkstats2
import thinkplot
import random
```

```python
rand_list = []
i = 0
while i < 1000:
    rand_list.append(random.random())
    i += 1

pmf = thinkstats2.Pmf(rand_list)
cdf = thinkstats2.Cdf(rand_list)
```

```python
thinkplot.Pmf(pmf, linewidth=0.1)
thinkplot.Config(xlabel='Random Num', ylabel='PMF')
```

```python
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel='Random Num', ylabel='CDF')
```

```python
print('Yes, the distribution of random numbers is uniform')
```
