```python
import thinkstats2
import pandas
import numpy as np
import statsmodels.formula.api as smf
import first
import nsfg
```

```python
live, firsts, others = first.MakeFrames()
live = live[live.prglngth>30]

resp = nsfg.ReadFemResp()
resp.index = resp.caseid

join_live_resp = live.join(resp, on='caseid', rsuffix='_r')
```

```python
model = smf.mnlogit('rmarital ~ age_r + C(race) + educat + totincr', data=join_live_resp)
results = model.fit()
results.summary()
```

```python
columns = ['age_r', 'race', 'educat', 'totincr']
new = pandas.DataFrame([[25, 2, 12, 11]], columns=columns)
y = results.predict(new)
y
```
